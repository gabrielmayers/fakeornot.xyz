```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df_train = pd.read_csv('https://raw.githubusercontent.com/gabrielmayers/fakeornot.xyz/master/Datasets/Original/train.csv')

df_train
```

# **Cleaning**: 

```
plt.figure(figsize=(16, 12))
sns.heatmap(df_train.isnull(), cmap='viridis', cbar=False)
```

```
df_train.columns
```

```
# Delete Null rows:

df_train_title = df_train[['title', 'label']]

df_train_title = pd.DataFrame(df_train_title)

df_train_title
```

```
plt.figure(figsize=(16, 12))
sns.heatmap(df_train_title.isnull(), cmap='viridis', cbar=False)
```

```
# Removing Null Title Values:

df_train_title.dropna(inplace=True)
```

```
plt.figure(figsize=(16, 12))
sns.heatmap(df_train_title.isnull(), cmap='viridis', cbar=False)
```

```
# Clearning test dataset:

df_test = pd.read_csv('https://raw.githubusercontent.com/gabrielmayers/fakeornot.xyz/master/Datasets/Original/test.csv')

df_test
```

```
plt.figure(figsize=(16, 12))
sns.heatmap(df_test.isnull(), cmap='viridis', cbar=False)
```

```
# Delete Null rows:

df_test_title = df_test['title']

df_test_title = pd.DataFrame(df_test_title)

df_test_title
```

```
# Removing Null Title Values:

df_test_title.dropna(inplace=True)
```

```
plt.figure(figsize=(16, 12))
sns.heatmap(df_test_title.isnull(), cmap='viridis', cbar=False)
```

```
# Saving: 

df_train_title.to_csv('train_to_model.csv')
df_test_title.to_csv('test_to_model.csv')
```
